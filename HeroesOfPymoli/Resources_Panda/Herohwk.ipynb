## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# number of Unique Items
Unique_Items = len(purchase_data_df["Item Name"].unique())
Unique_Items
# average price
avg_price = purchase_data_df["Price"].mean()
avg_price
# number of purchases
purchases = len(purchase_data_df)
purchases
# Total Revenue
Revenue = purchase_data_df["Price"].sum()
Revenue
# Summary Tables
Summary_dict = {"Unique Items":[Unique_Items],
             "Average Price":[avg_price],
             "Total Purchases":[purchases],
             "Total Revenue":[Revenue]}
summary_df = pd.DataFrame(Summary_dict)

summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Purchases"] = summary_df["Total Purchases"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

## Age Demographics

Purchasing Analysis (Age)
Bin the purchase_data data frame by age
Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
Create a summary data frame to hold the results
Optional: give the displayed data cleaner formatting
Display the summary data frame


```python
#create the bins
agemax = purchase_data_df["Age"].max()
print(agemax)
agemin = purchase_data_df["Age"].min()
print(agemin)

bins = [0, 9, 14, 19, 24, 29, 34, 39, 75]
group_names = ["<10", "10-14", "15-19","20-24","25-29", "30-34", "35-39", "40+"]
# df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names, include_lowest=True)
# df
purchase_data_df["Age Item"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
purchase_data_df["Age Item"]
#group the data by age group
agegroup = purchase_data_df.groupby(["Age Item"])
#print(agegroup)
#purchase count of each age group
purcount = agegroup["SN"].nunique()
#print(purcount)
#percentage of each age group
totpurcount = len(purchase_data_df["SN"])
#print(totpurcount)
percent= round(((purcount/totpurcount)*100),2)
percent
#Average Purchase Price
avgagepur = round((purcount/totpurcount)*100,2)
#print(avgagepur)
#Total Purchase Price
totpurc = agegroup["Price"].sum()
#print(totpurc)
#Average Purchase Total per Person by age group.
kdkd = round(totpurc/purcount, 2)
#print(kdkd)
Summary_AgeTable= pd.DataFrame()
Summary_AgeTable = {"Purchase Count":purcount,
                   "Average Price":avgagepur,
                   "Total Purchase":totpurc,
                   "Average Total per Person":kdkd}
Summary_AgeTable = pd.DataFrame(Summary_AgeTable)
Summary_AgeTable
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

#

```python
#Gather new columns
NewGroup= purchase_data_df[["Item ID", "Item Name", "Price"]]
NewGroup
# #select only the item group
items = NewGroup.groupby(["Item ID","Item Name"])
items
# #purchase count item
itempcount = items["Price"].count()
itempcount
# #total purchase value
totime  = (items["Price"].sum()) 
totime
#item prices
avgitemprice= totime/itempcount
avgitemprice
#most popular
Popular = pd.DataFrame({"Purchase Count": itempcount, 
                        "Item Price":  avgitemprice,
                       "Total Purchase Value": totime})
Popular
Popular["Item Price"] = Popular["Item Price"].map("${:,.2f}".format)
Popular["Total Purchase Value"] = Popular["Total Purchase Value"].map("${:,.2f}".format)
#most popular
TopItems = Popular.sort_values(["Purchase Count"], ascending=False)
TopItems
```
